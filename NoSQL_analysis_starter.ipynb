### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}
results = list(establishments.find(query))
# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print("Number of documents in result:", count)
# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(len(hygiene_df))

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
# ChatGPT
rating_query = {'RatingValue': {'$gte': 4}, '$or': [{'AddressLine3': 'London'}, {'AddressLine4': 'London'}]}
rating_results = list(establishments.find(rating_query))

# Use count_documents to display the number of documents in the result
rating_count = establishments.count_documents(rating_query)
print("Number of documents in result:", rating_count)
# # Display the first document in the results using pprint
pprint(rating_results[0])
```

```python
# Convert the result to a Pandas DataFrame
rating_df = pd.DataFrame(rating_results)

# Display the number of rows in the DataFrame
print(len(rating_df))

# Display the first 10 rows of the DataFrame
rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
# ChatGPT

from decimal import Decimal
from bson.decimal128 import Decimal128

penang_query = {"BusinessName": "Penang Flavours"}

penang_record = establishments.find_one(penang_query)

lon_value = penang_record['geocode']['longitude'].to_decimal()
lat_value = penang_record['geocode']['latitude'].to_decimal()

degree_search = 0.01
degree_decimal = Decimal(str(degree_search))

latitude1 = Decimal128(lat_value + degree_decimal)
latitude2 = Decimal128(lat_value - degree_decimal)
longitude1 = Decimal128(lon_value + degree_decimal)
longitude2 = Decimal128(lon_value - degree_decimal)


geo_query = {'RatingValue': {'$eq': 5}, 'geocode.longitude': {'$gte': longitude2},\
 'geocode.longitude': {'$lte': longitude1}, 'geocode.latitude': {'$gte': latitude2},\
 'geocode.latitude': {'$lte': latitude1}}
sort =  [('scores.Hygiene', 1)]
limit = 5

geo_results = list(establishments.find(geo_query).sort(sort).limit(limit))
# Print the results
pprint(geo_results)
```

```python
# Convert result to Pandas DataFrame
geo_df = pd.DataFrame(geo_results)
geo_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}

group_query = {'$group': {'_id': {"LocalAuthorityName": "$LocalAuthorityName"}, 
                          'count': { '$sum': 1 }
                         }
              }

sort_values = {'$sort': { 'count': -1 }}

pipeline = [match_query, group_query, sort_values]

authority_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in result: ", len(authority_results))

# Print the first 10 results
pprint(authority_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
authority_df = pd.DataFrame(authority_results)

# Display the number of rows in the DataFrame
print(len(authority_df))

# Display the first 10 rows of the DataFrame
authority_df.head(10)
```
